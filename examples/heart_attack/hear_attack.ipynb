```python
import os
import sys
from pathlib import Path
root_path = Path.cwd().parent.parent
sys.path.append(root_path)

file_path = root_path / "examples/heart_attack/heart_attack_dataset.csv"
```

### Load without transformer

```python
from gaubongai.data_management import DataProcessor
from gaubongai.data_management.loaders import PandasCSVLoader
from gaubongai.data_management.transformers import PandasDfTransformer

loader = PandasCSVLoader()
data_processor = DataProcessor(loader=loader)

data_container = data_processor.process_file(file_path)
data_container.data.head()
```

### Load with transformer

```python
loader = PandasCSVLoader()
transformer = PandasDfTransformer(
    rename_columns={
        "Age": "age",
    }
)

data_processor = DataProcessor(loader=loader, transformers=[transformer])

data1 = data_processor.process_file(file_path)
data1.data.head()
```

### Transform data container

```python
transformer2 = PandasDfTransformer(
    rename_columns={

        "Gender": "gender",

    }
)
data_processor2 = DataProcessor(transformers=[transformer2])
data2 = data_processor2.process_data_container(data1)
data2.data.head()
```
